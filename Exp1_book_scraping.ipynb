```python
import requests
import csv
import pandas as pd
from bs4 import BeautifulSoup

def scrap_book():
    all_book_details=[]

    for i in range(1,51):
        api_url=f"http://books.toscrape.com/catalogue/category/books_1/page-{i}.html"

        headers = {"User-Agent" : "Mozilla/5.0"}

        response=requests.get(api_url,headers=headers)

        if response.status_code==200:
            
            soup = BeautifulSoup(response.text, 'html.parser')
            books = soup.find('ol').find_all('article', class_='product_pod')

            for book in books:
                title=book.h3.a['title']
                price_sign=book.find('p', class_='price_color').text.strip()
                price = float(price_sign.replace('Â£', '').strip())
                
                ratings=book.p['class'][1]

                book_info={
                    "Title": title,
                    "Price": price,
                    "Ratings": ratings
                }

                all_book_details.append(book_info)

            df=pd.DataFrame(all_book_details)

            df.to_csv("book_details.csv",index=False)
            print("All the data saved to 'book_details.csv'")
            print(df.head())

        else:
            print("Failed to retrieve data. Status code: ",response.status_code)

scrap_book()
```
